As was said in the seminar, it is possible to interpret the task in two ways. 
First, we can try to answer **how many times** we want to roll before the rolling. It is more difficult situation because we don't know exact realization of first several rolls.

Secondly, we can try to answer should we stop **if we know what is the value of current sum $s_{k}$**. In this situation, we can calculate expected gain given $s_k$ and stop on maximal $s_k$ for which expected gain is non-negative. 

We will cover the second case.

### Exercise 1. 

So the possible words are
tuples $(Seq,T)$ where $Seq \in \{1,2,3,4,5,6\}^k $ and $T \in \{Loaded, Fair\}$

So given that $$s_k = \sum_{i=1}^k h_i,$$
where $h_k$ is outcome of the one rolling, we can write down outcome of our game as 
$$R_k =  s_k  X \cdot \mathbb{1}[s_k\leq 20].$$

 

### Exercise 2.
For all $s_k \leq 14$ expected gain is equal to $\mathbb{E}[h_k] =3,5>0$

For $s_k = 15$ we can find conditional expectation our the $R_{k+1}$. 

$\mathbb{E} [R_{k+1}|s_k=15] = \frac{1}{6}X(16+17+18+19+20) = 15X$
So for the case when $s_k=15$ the expected gain in 0 so we can roll or stop (e.g. any of these dexisions whould be rational and the exact choice depend on risk polisy of the player)

$\mathbb{E} [R_{k+1}|s_k=16] = \frac{1}{6}X(17+18+19+20) \leq 16X.$ 

So we should stop rolling if we've got $s_k \geq 16$.


### Exercise 3. 

The same idea as Ex.2, however, the die is loaded.

For all $s_k \leq 14$ expected gain is equal to $\mathbb{E}[h_k] =3,5>0$

For $s_k = 15$ we can find conditional expectation our the $R_{k+1}$. 
$\mathbb{E} [R_{k+1}|s_k=15] = X\left(\frac{1}{12}(15+1+15+2+15+3)+\frac{1}{12}(15+4+15+5+0)\right) = 16X \geq 15X $
As our expacted gain is positive $16X - 15X = X \geq 0 $, we should roll more. 

$\mathbb{E} [R_{k+1}|s_k=16] = X(\frac{17+18+19}{12}+\frac{20}{4}) \leq 16X.$ 

So we should stop rolling if we've got $s_k \geq 16$.

### Exercise 4. 
We have sequence $S_1 = (1,2,6,6,4)$ and want to find its probability.
$$P(S_1) = P(S_1|T = Loaded)P(T = Loaded)+ P(S_1|T = Fair)P(T = Fair).$$
Now let us use the fact that rolls in our siquence are independent given that we know the type of the dice. $$P(S_1|T = Loaded) = \prod_{i=1}^{5}P(h_i=s_i|T = Loaded)$$
$$P(S_1) = {\left(\frac{1}{4}\right)}^2{\left(\frac{1}{12}\right)}^3 \frac{1}{2}+{\left(\frac{1}{6}\right)}^5 \frac{1}{2} = 8.24 \cdot 10^{-5}$$

Using the same conditional independence we will get the same probability for any permutation of the original sequence, so $P(S_1)=P(S_2)= P(S_3)$.
### Exercise 5. 
We can use Bayes rule for finding conditional probability that the dice are fair given the sequence. 
 
$$P(T = Fair | S_1) = \frac{P(S_1| T = Fair)}{P(S_1)} = \frac{{\left(\frac{1}{6}\right)}^5 \frac{1}{2}}{8.24 \cdot 10^{-5}} =0.78 $$

### Exercise 7.

If we will look at expected gain given any sequence $S = (h_i)_{i=1}^k$  the value of $$\mathbb{E}[R_{k+1}|s_k=15, S ] = \mathbb{E}[R_{k+1}|s_k=15, T = Loaded ] P(T = Loaded | S) \\+\mathbb{E}[R_{k+1}|s_k=15, T = Fair ] P(T = Fair | S) = 15 Xp + 16X(1- p) \geq 15X$$.

Whereas

The values $\mathbb{E}[R_{k+1}|s_k=15, T = Fair ]$  and $\mathbb{E}[R_{k+1}|s_k=15, T = Loaded ]$ are from Exercise 2 and 3.
